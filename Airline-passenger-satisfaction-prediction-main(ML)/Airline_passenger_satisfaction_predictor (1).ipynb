```python
import pandas as pd
import numpy as np
```

```python
X=pd.read_csv("train.csv")
X_train=X.iloc[:,:-1]
y_train=X.iloc[:,-1:]
```

```python
y=pd.read_csv("test.csv")
X_test=y.iloc[:,:-1]
y_test=y.iloc[:,-1:]
```

```python
y_test = pd.get_dummies(y_test)
y_test=y_test['satisfaction_satisfied']
#y_test=pd.DataFrame(y_test)
y_test
```

```python
y_train = pd.get_dummies(y_train)
y_train=y_train['satisfaction_satisfied']
#y_train=pd.DataFrame(y_train)
y_train
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.metrics import accuracy_score
```

```python
X_train=X_train.drop(["Type of Travel","Class","Gender","id","Unnamed: 0"],axis=1)
X_ones=pd.get_dummies(X_train["Customer Type"])["Loyal Customer"]
X_train=pd.concat((X_ones,X_train),axis=1)
X_train=X_train.drop(["Customer Type"],axis=1)
X_train.fillna(X_train.mean(),inplace=True)
X_train
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_train
```

```python
X_test
```

```python
X_test=X_test.drop(["Type of Travel","Class","Gender","id","Unnamed: 0"],axis=1)
X_ones=pd.get_dummies(X_test["Customer Type"])["Loyal Customer"]
X_test=pd.concat((X_ones,X_test),axis=1)
X_test=X_test.drop(["Customer Type"],axis=1)
X_test.fillna(X_test.mean(),inplace=True)
X_test
```

```python
X_test = scaler.transform(X_test)
X_test
```

```python
y_test
```

```python
X_test
```

```python
model=Sequential()
model.add(tf.keras.layers.Dense(units=10,activation='relu',input_dim=len(X_train[0])))
model.add(tf.keras.layers.Dense(units=10,activation='relu'))
model.add(tf.keras.layers.Dense(units=1,activation='sigmoid'))
```

```python
from tensorflow.keras.optimizers import Adagrad
cust=Adagrad(learning_rate=0.1)
model.compile(loss='binary_crossentropy',optimizer=cust ,metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=100, batch_size=600)
```

```python
y_pred=model.predict(X_test)
y_pred= [0 if val<0.5 else 1 for val in y_pred]
```

```python
accuracy_score(y_test,y_pred)
```
